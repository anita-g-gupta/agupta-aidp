# Import libraries

```python
from pyspark.sql.types import *
from pyspark.sql.functions import *
```

# Define Schema

```python
from pyspark.sql.types import *

# Customer Schema
customer_schema = StructType([
    StructField("customer_id", IntegerType(), True),
    StructField("customer_name", StringType(), True),
    StructField("age", IntegerType(), True),
    StructField("annual_income", DoubleType(), True),
    StructField("employment_status", StringType(), True),
    StructField("credit_score", IntegerType(), True)
])

# Loan Schema
loan_schema = StructType([
    StructField("loan_id", IntegerType(), True),
    StructField("customer_id", IntegerType(), True),
    StructField("loan_amount", DoubleType(), True),
    StructField("loan_type", StringType(), True),
    StructField("interest_rate", DoubleType(), True),
    StructField("loan_term_months", IntegerType(), True),
    StructField("collateral_value", DoubleType(), True)
])

# Payment Transactions Schema
payment_schema = StructType([
    StructField("payment_id", IntegerType(), True),
    StructField("loan_id", IntegerType(), True),
    StructField("payment_amount", DoubleType(), True),
    StructField("days_past_due", IntegerType(), True),
    StructField("payment_status", StringType(), True)
])

# Customer Financials Schema
financials_schema = StructType([
    StructField("customer_id", IntegerType(), True),
    StructField("total_debt", DoubleType(), True),
    StructField("equity", DoubleType(), True),
    StructField("revenue", DoubleType(), True),
    StructField("revenue_decline", IntegerType(), True)
])

# Macro Market Schema
macro_schema = StructType([
    StructField("record_id", IntegerType(), True),
    StructField("date", DateType(), True),
    StructField("interest_rate", DoubleType(), True),
    StructField("inflation_rate", DoubleType(), True),
    StructField("unemployment_rate", DoubleType(), True)
])
```

# Define File Paths

```python
customer_path = "/Volumes/credit_risk_assessment/bronze/raw_data/customers.csv"

loan_path = "/Volumes/credit_risk_assessment/bronze/raw_data/loans.csv"

payment_path = "/Volumes/credit_risk_assessment/bronze/raw_data/payments.csv"

financials_path = "/Volumes/credit_risk_assessment/bronze/raw_data/financials.csv"

macro_path = "/Volumes/credit_risk_assessment/bronze/raw_data/macro_market.csv"
```

# Read CSV Files Using Schema

```python
customers_df = spark.read.format("csv") \
    .option("header","true") \
    .schema(customer_schema) \
    .load(customer_path)

loans_df = spark.read.format("csv") \
    .option("header","true") \
    .schema(loan_schema) \
    .load(loan_path)

payments_df = spark.read.format("csv") \
    .option("header","true") \
    .schema(payment_schema) \
    .load(payment_path)

financials_df = spark.read.format("csv") \
    .option("header","true") \
    .schema(financials_schema) \
    .load(financials_path)

macro_df = spark.read.format("csv") \
    .option("header","true") \
    .schema(macro_schema) \
    .load(macro_path)
```

# Quick Validation

```python
customers_df.show(3)
loans_df.show(2)
payments_df.show(2)
financials_df.show(2)
macro_df.show(2)
```

# Write to Bronze Delta Tables

```python
customers_df.write.mode("overwrite").saveAsTable("credit_risk_assessment.bronze.customers")

loans_df.write.mode("overwrite").saveAsTable("credit_risk_assessment.bronze.loans")

payments_df.write.mode("overwrite").saveAsTable("credit_risk_assessment.bronze.payment_transactions")

financials_df.write.mode("overwrite").saveAsTable("credit_risk_assessment.bronze.customer_financials")

macro_df.write.mode("overwrite").saveAsTable("credit_risk_assessment.bronze.macro_market")
```
